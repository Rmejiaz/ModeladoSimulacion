## 6.4 Rabbit versus Sheep

Consider the following "Rabbit versus Sheep" problems, where $x,y \geq 0$. Find the fixed points, investigate their stability, draw the nullclines, and sketch plausible phase portraits. Indicate the basins of attraction of any stable fixed points.

```python
# Funcion para graficar el retrato de fase de cualquier sistema

def plot_phase_portrait(fun):
  t_span = [0,15]
  plt.figure(figsize=(16,9))

  for x0 in np.arange(0,3,0.3):
    for y0 in np.arange(0,3,0.3): 
        y_0 = [x0,y0]
        t = np.arange(0,15,0.01)
        sol = solver(fun, t_span = t_span, y0 = y_0,t_eval=t)

        plt.plot(sol.y[0],sol.y[1])
        plt.title("Phase portrait",fontsize=20)
        plt.xlabel("$x$",fontsize=20)
        plt.ylabel("$y$",fontsize=20)
  plt.show()
```

### 6.4.1

$\dot x = x(3-x-y)$

$\dot y = y(2-x-y)$

```python
def fun1(t,y_):
  x,y = y_
  x_dot = x*(3-x-y)
  y_dot = y*(2-x-y)
  return x_dot, y_dot

plot_phase_portrait(fun1)
```

### 6.4.2

$\dot x = x(3-2x-y)$

$\dot y = y(2-x-y)$

```python
def fun2(t,y_):
  x,y = y_
  x_dot = x*(3-2*x-y)
  y_dot = y*(2-x-y)
  return x_dot, y_dot

plot_phase_portrait(fun2)
```

### 6.4.3

$\dot x = x(3-2x-2y)$

$\dot y = y(2-x-y)$

```python
def fun3(t,y_):
  x,y = y_
  x_dot = x*(3-2*x-2*y)
  y_dot = y*(2-x-y)
  return x_dot, y_dot

plot_phase_portrait(fun3)
```
