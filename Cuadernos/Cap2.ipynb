```
# Importamos las librerías

import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display, Math
```

```
### Definición de funciones

# Función del método de euler

def euler_method(f, h, x0, ti, tf):
  N = int((tf-ti)/h)+1
  x = np.zeros(N)
  x[0] = x0
  for i in range(len(x)-1):
    x[i+1] = x[i] + h*f(x[i])

  t = np.linspace(ti,tf,N)
  
  return t,x

# Función para graficar el phase portrait y el vector field

def plot_phase_portrait(f,step,ti,tf,x0_i,x0_f,n):
  
  plt.figure(figsize=(20,9))

  plt.subplot(121)
  for i in np.linspace(x0_i,x0_f,n):
    x0 = i
    t,x = euler_method(f,step,x0,ti,tf)
    plt.plot(t,x,label=f'$x_0 = {i}$')
    plt.xlabel("$t$",fontsize=15)
    plt.ylabel("$x(t)$",fontsize=15)
    plt.title("Phase Portrait",fontsize=15)
    # plt.legend()
  plt.subplot(122)
  x_plot=np.linspace(x0_i,x0_f,100)
  plt.plot(x_plot,f(x_plot))
  plt.xlabel("$x$",fontsize=15)
  plt.ylabel("$\dot x$",fontsize=15)
  plt.title("Vector Field",fontsize=15)
  plt.show()
```

b)

Ahora, utilizando el método de Euler:

```
from IPython.display import display, Math

def func11(x):
  return -x

step = 1 # delta t
x0 = 1 # condición inicial x(0) = 1
ti = 0 # tiempo inicial
tf = 1 # tiempo final x(1)

for n in range(5):
  step = 10**(-n)
  t,x = euler_method(func11,step,x0,ti,tf)
  display(Math(f'\Delta t = {step} \\rightarrow \hat x = {x[-1]}')) # interesa conocer x(1), es decir el ultimo elemento del array
```

c)

```
def func11(x):
  return -x

step = 1 # delta t
x0 = 1 # condición inicial x(0) = 1
ti = 0 # tiempo inicial
tf = 1 # tiempo final x(1)

real_solution = np.exp(-1)
deltas = np.array([])
euler_errors = np.array([])

for n in range(5):
  step = 10**(-n)
  deltas = np.append(deltas,step)
  t,x = euler_method(func11,step,x0,ti,tf)
  euler_errors = np.append(euler_errors,abs(x[-1] - real_solution))

plt.figure(figsize=(16,9))
plt.subplot(121)
plt.plot(deltas,euler_errors)
plt.title("$E$ as a function of $\Delta t$",fontsize=15)
plt.xlabel("$\Delta t$",fontsize=15)
plt.ylabel("E",fontsize=15)

plt.subplot(122)
plt.plot(np.log(deltas),np.log(euler_errors))
plt.title("$\ln (E)$ as a function of $\ln (\Delta t)$",fontsize=15)
plt.xlabel("$\ln \Delta t$",fontsize=15)
plt.ylabel("$\ln (E)$",fontsize=15)


plt.show()
```

Como se puede ver, a medida que se aumenta el tamaño del paso, el error crece, pues la aproximación se hace cada vez más mala. Además, se puede ver que el error crece (o decrece) de forma lineal conforme se cambia el tamaño del paso. En pocas palabras, podemos concluir que el error y el tamaño del paso son directamente proporcionales.

$E \propto \Delta t$

### 2.8.4
 
 Redo Exercise 2.8.3, using improved Euler method

```
### Improved euler method:

def improved_euler(f,h,x0,ti,tf):
  N = int((tf-ti)/h)+1
  x = np.zeros(N)
  x[0] = x0
  for i in range(len(x)-1):
    trial_step  = x[i] + h*f(x[i])
    x[i+1] = x[i] + 0.5*(f(x[i]) + f(trial_step))*h
    
  t = np.linspace(ti,tf,N)
  
  return t,x
```

```
plt.figure(figsize=(16,9))
plt.subplot(121)
plt.plot(deltas,improved_errors,label='Improved Euler Method')
plt.plot(deltas, euler_errors,label='Euler Method')
plt.title("$E$ as a function of $\Delta t$",fontsize=15)
plt.xlabel("$\Delta t$",fontsize=15)
plt.ylabel("E",fontsize=15)

plt.legend()

plt.subplot(122)
plt.plot(np.log(deltas),np.log(improved_errors),label='Improved Euler Method')
plt.plot(np.log(deltas),np.log(euler_errors),label = 'Euler Method')
plt.title("$\ln (E)$ as a function of $\ln (\Delta t)$",fontsize=15)
plt.xlabel("$\ln \Delta t$",fontsize=15)
plt.ylabel("$\ln (E)$",fontsize=15)

plt.legend()

plt.show()
```

### 2.8.5

Redo Exercise 2.8.3, using Runge-Kutta method.

```
### Runge-Kutta Method:

def runge_kutta(f,h,x0,ti,tf):
  N = int((tf-ti)/h)+1
  x = np.zeros(N)
  x[0] = x0
  for i in range(len(x)-1):
    k1 = f(x[i])*h
    k2 = f(x[i] + 0.5*k1)*h
    k3 = f(x[i] + 0.5*k2)*h
    k4 = f(x[i] + k3)*h

    x[i+1] = x[i] + (k1 + 2*k2 + 2*k3 + k4)/6
    
  t = np.linspace(ti,tf,N)
  
  return t,x
```
